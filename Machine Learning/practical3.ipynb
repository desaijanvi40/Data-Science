```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

#label encoding used when we can set some order in dataset
data={
"Employee_ID":[1,2,3,4,5],
"Department":["HR","IT","Finanace","IT","HR"],
"Education_Level":["Bachlor","Master","PhD","High School","Master"],
"Performance_Rating":["high","Medium","Low","Medium","Low"],
"Experience":[2,5,8,1,4],
"Salary":[25000,50000,80000,20000,45000]    

}

df=pd.DataFrame(data)

print("orignal dataset is \n")
print(df)

educational_order={
    "High School":1,
    "Bachlor":2,
    "Master":3,
    "PhD":4
}

performance_order={
    "Low":1,
    "Medium":2,
    "high":3
}

df["Education_Ordinal"]=df["Education_Level"].map(educational_order)
df["Performance_Ordinal"]=df["Performance_Rating"].map(performance_order)

print("After Label Encoding dataset is \n")
print(df)
```
