```python
# One Hot Encoding and Dummy Variables

import pandas as pd

data={
"Customer_ID":[1,2,3,4,5],
"Gender":["Male","Female","Female","Male","Female"],
"City":["Ahmedabad","Mumbai","Delhi","Mumbai","Ahmedabad"],
"Subscription_Type":["Basic","Premium","Standard","Basic","Premium"],
"Monthly_Spending":[500,1500,300,7000,400]    
}

df=pd.DataFrame(data)
print("Orignal Dataset is \n")
print(df)

#one Hot Encoding
one_hot_df=pd.get_dummies(df,columns=["Gender","City","Subscription_Type"],dtype=int)

print("Data set after one hot encoding")
print(one_hot_df)

#drop_first=true
dummy_df=pd.get_dummies(df,columns=["Gender","City","Subscription_Type"],drop_first=True,dtype=int)


print("Dataset after Dummy varibles (k-1) Encoding: \n")
print(dummy_df)
```
