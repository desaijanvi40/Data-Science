```python
import numpy as np
import pandas as pd

data={
    "Student_ID":[101,102,103,104,105],
    "Name":["kashyap","ved","jay","kunal","chirag"],
    "Attendance":[50,np.nan,85,88,np.nan],
    "Study_Hours":[4,5,np.nan,6,3],
    "Assignment_Marks":[18,np.nan,15,19,17],
    "Internal_Marks":[20,18,17,19,16],
    "Final_Marks":[78,82,75,88,70]
    
}

# 1)Load the dataset into Pandas.
df=pd.DataFrame(data)

print(f"orignal dataset is:\n ")
print(df)


# 2)Identify which columns contain missing values.
# 3)Count total missing values column-wise.

print("missing values count:\n")
print(df.isnull().sum())

# 4)Calculate percentage of missing values in each column.
print("missing value percentage \n")
print(df.isnull().sum()/len(df) *100)

# 5)fill



df

```

```python
df["Attendance"].fillna(df["Attendance"].median(),inplace=True)
```

```python
df["Study_Hours"].fillna(df["Study_Hours"].mean(),inplace=True)
```

```python
df["Assignment_Marks"].fillna(df["Assignment_Marks"].mode())
```

```python
print(f"after cleaning dataset is:\n ")
print(df)
```
