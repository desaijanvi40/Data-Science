```python
import pandas as pd
import matplotlib.pyplot as plt

data={

    "Employee_ID":[1,2,3,4,5,6],
    "Monthly_Salary":[25000,30000,32000,35000,5000000,31000]
}

df=pd.DataFrame(data)
print("orignal dataset is \n")
print(df)

#boxplot
plt.figure(figsize=(6,4))
plt.boxplot(df["Monthly_Salary"])
plt.title("boxplot before outlier removal")
plt.show()

#IQR
Q1=df["Monthly_Salary"].quantile(0.25)
Q3=df["Monthly_Salary"].quantile(0.75)

IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR

print("Lower bound",lower_bound)
print("Upper bound",upper_bound)
```

```python
#remove outlier
df_cleaned=df[(df["Monthly_Salary"] >= lower_bound) & (df["Monthly_Salary"]<=upper_bound)]

print("after outlier removal dataset is \n")
print(df_cleaned)

#boxplot
plt.figure(figsize=(6,4))
plt.boxplot(df_cleaned["Monthly_Salary"])
plt.title("boxplot after outlier removal")
plt.show()
```
