```python
import pandas as pd
import numpy as np

# Create Population Dataset (1000 Students)
np.random.seed(42)

population_marks = np.random.normal(loc=70, scale=10, size=1000)

df_population = pd.DataFrame(population_marks, columns=["Marks"])

df_population.head()



# Calculate Population Mean & Variance

population_mean = df_population["Marks"].mean()
population_variance = df_population["Marks"].var()


print("Population Mean:", population_mean)
print("Population Variance:", population_variance)

# # Take Random Sample (100 Students)
# sample_df = df_population.sample(n=100, random_state=1)

# sample_mean = sample_df["Marks"].mean()
# sample_variance = sample_df["Marks"].var()


```

```python
# Salary Dataset Example
np.random.seed(100)

salary_population = np.random.normal(loc=50000, scale=8000, size=5000)

salary_df = pd.DataFrame(salary_population, columns=["Salary"])

salary_df.head()

true_mean = salary_df["Salary"].mean()
print("True Population Salary Mean:", true_mean)

random_sample = salary_df.sample(n=200, random_state=10)

print("Random Sample Mean:", random_sample["Salary"].mean())

biased_sample = salary_df[salary_df["Salary"] > 55000].sample(n=200, random_state=10)

print("Biased Sample Mean:", biased_sample["Salary"].mean())
```
