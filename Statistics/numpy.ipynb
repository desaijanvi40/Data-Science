```python
#copy vs view
# diffrence is copy method create in array inside memory block , but view methods refers or 
#     points to orignal array
a2=np.array([1,2,3,4])
copy_arr=np.copy(a2)#1 2 3 4

print(a2,copy_arr)
copy_arr[2]=89
print("copy",copy_arr)
print("orignal",a2)

view_arr=a2.view()
print(a2,view_arr)
view_arr[2]=89
print("view",view_arr)
print("ori",a2)
```
