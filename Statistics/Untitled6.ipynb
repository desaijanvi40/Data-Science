```python
ages=[23,24,24,32,32,43,12,43,67,45,32]

import numpy as np


#outlier effects on mean
print(np.mean(ages))
ages=np.sort(ages)
print(ages)

#but outlier will not effect to the median
print(np.median(ages))

#pip install statistics
import statistics as st
print(st.mode(ages))
```

```python
#pip install seaborn
import seaborn as sns

sns.boxplot(ages)
```

```python
#5 number summary
d_min=np.min(ages)

Q1=np.percentile(ages,25)

median=np.median(ages)

Q3=np.percentile(ages,75)

d_max=np.max(ages)

print(f"min is {d_min} , Q1 is {Q1} , median is {median} , Q3 is {Q3} ,  max is {d_max}")
```

```python
# how to identify outlier

IQR=Q3-Q1

lower=Q1-(1.5*IQR)
higher=Q3+(1.5*IQR)

print(f"outside of this range is outlier range:{lower} - {higher}")
```

```python
#variance

#population data division /n
print(np.var(ages,axis=0))

# division n-1
print(st.variance(ages))

import math
std=math.sqrt(np.var(ages,axis=0))
print(std)
```

```python
#histogram

sns.histplot(ages,kde=True)
```
