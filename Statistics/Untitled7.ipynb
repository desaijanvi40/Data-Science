```python
import numpy as np

rolls=np.random.randint(1,7,1000)
# 5 3 6 2 1 

even_count=sum(rolls%2 == 0)

print(even_count)

probability=even_count/1000

print(f"probability is {probability}")
```

```python
toss=np.random.choice(["head","tail"],25)

print(toss)

```

```python
sum_head=sum(toss == "head")

print(sum_head)

probability=sum_head/25

print(f"probability of coming head is {probability}")
```

```python
# 1000 time
# 1st time coin toss and roll dice

# how many chances of coming head and even number together

coin=np.random.choice(["h","t"],1000)

dice=np.random.randint(1,7,1000)

count=0

for i in range(1000):
    if coin[i] == "h" and dice[i]%2 == 0:
        count=count+1

print(f"total count is {count}")    


probability=count/1000

print(f"probability of coming head and even number on dice together is {probability}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.binomial(n=10,p=0.5,size=1000)

# print(data)

plt.hist(data)
plt.title("Binomial Distribution")
plt.show()
```
